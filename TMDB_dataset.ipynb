```python
import pandas as pd
```

```python
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=a4bd454e03e9300771782a8bf6268b2f&page=1&sort_by=created_at.asc')
```

```python
response.json()
```

```python
df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
```

```python
df.head()
```

```python
df  = pd.DataFrame()
```

```python
df
```

```python
for i in range(1,502):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=a4bd454e03e9300771782a8bf6268b2f&page={}&sort_by=created_at.asc'.format(i))
  if 'results' in response.json():
        temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        df = pd.concat([df, temp_df], ignore_index=True)
  else:
        print(f"Warning: 'results' key not found in response for page {i}. Skipping this page.")
```
